```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
#reading dataset
df_train = pd.read_csv('C:/A_stuff/Learning/Machine Learning/Kaggle/Titanic Problem/train.csv')
df_test = pd.read_csv('C:/A_stuff/Learning/Machine Learning/Kaggle/Titanic Problem/test.csv')
```

```python
#Train data visualization
df_train.head(5)
```

```python
df_test.head()
```

```python
df_train.shape
```

```python
#Age, Cabin and Embarked are having missing value
#treating missing values
df_train.info()
```

```python
#Missing values-Age
df_train['Age'].describe()

#lets substitute it by mean
def impute_age(df):
    df['Age']= df['Age'].fillna(df['Age'].mean())
    return df
       
df_train = impute_age(df_train)
df_test = impute_age(df_test)
```

```python
#Missing values-Cabin
df_train['Cabin'].mode()[0]

#lets substitute it by mode
def impute_cabin(df):
    df['Cabin']= df['Cabin'].fillna(df_train['Cabin'].mode()[0])
    return df
       
df_train = impute_cabin(df_train)
df_test = impute_cabin(df_test)
```

```python
#Missing values- Embarked
df_train['Embarked'].mode()[0]

#lets substitute it by mode
def impute_cabin(df):
    df['Embarked']= df['Embarked'].fillna(df_train['Embarked'].mode()[0])
    return df
       
df_train = impute_cabin(df_train)
df_test = impute_cabin(df_test)
```

```python
df_train.head(5)
```

```python
#lets delete Passenger ID from it as it is of now use
def delete_passengerID(df):
    df = df.drop(['PassengerId'],axis =1)
    return df
       
df_train = delete_passengerID(df_train)
df_test = delete_passengerID(df_test)
```

```python
df_train.head(5)
```

```python
#lets delete Name,Ticket,Fare,Cabin 
def delete_columns(df):
    df = df.drop(['Name','Ticket','Fare','Cabin'],axis =1)
    return df
       
df_train = delete_columns(df_train)
df_test = delete_columns(df_test)
```

```python
df_train.head(5)
```

```python
#label encoding for Sex and embarked

def label_encoding(df,col):
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    return le,df


le_sex,df_train = label_encoding(df_train,'Sex')
df_test['Sex'] = le_sex.transform(df_test['Sex'])

le_embarked,df_train = label_encoding(df_train,'Embarked')
df_test['Embarked'] = le_embarked.transform(df_test['Embarked'])
```

```python
df_train.head(5)
df_test.head(5)
```

```python
#one hot encoding for train and test
#Pclass and Embarked
def create_dummies(df,col):
    df = pd.concat([df,pd.get_dummies(df[col],drop_first=True,prefix=col)],axis =1)
    df = df.drop([col],axis =1)
    return df

df_train = create_dummies(df_train,'Pclass')
df_test = create_dummies(df_test,'Pclass')


df_train = create_dummies(df_train,'Embarked')
df_test = create_dummies(df_test,'Embarked')

```

```python
df_train.head(5)
```

```python
#creating X and Y matrix
Y = df_train.iloc[:,0]
X = df_train.iloc[:,1:]
```

```python
#dividing it into train and test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
#Model 
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

#classifier = LogisticRegression()
classifier = GaussianNB()
```

```python
#Fitting model and making prediction
classifier.fit(X_train,Y_train)
Y_predict = classifier.predict(X_test)
```

```python
#Model Evaluation
from sklearn.model_selection import cross_val_score
print('Train score ',cross_val_score(classifier,X_train,Y_train,cv=3,scoring = 'accuracy'))
print('Test score ',cross_val_score(classifier,X_test,Y_test,cv=3,scoring = 'accuracy'))
```
